```python
import matplotlib.pyplot as plt

#dados['tempo'].value_counts().plot.bar(color=['pink', 'aqua', 'silver'])

fig = plt.figure()

ax = fig.add_axes([0,0,1,2])

plt.title('Tempo de Teste', fontweight='bold', fontsize='15') 
plt.xlabel('Qnt de Horas', fontweight='light', fontsize='14')
plt.ylabel('C√©lulas que Reagiram', fontweight='light', fontsize='14')


dados_temp = dados['tempo'].value_counts()

a = dados_temp.loc[48]
b = dados_temp.loc[72]
c = dados_temp.loc[24]

ax.bar("48hrs", a, label="Rea√ß√£o em 48 Hrs", color="pink")
ax.bar("72hrs", b, label="Rea√ß√£o em 72 Hrs", color="aqua") 
ax.bar("24hrs", c, label="Rea√ß√£o em 24 Hrs", color="silver") 

plt.legend()

```
