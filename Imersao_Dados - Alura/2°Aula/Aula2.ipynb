```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize = (8, 6))

ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'), order = dados['composto'].value_counts().index[0:5])
ax.set_title('Top 5 Compostos')

plt.show()
```

###Desafio 02:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize = (8, 6))

ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'), order = dados['composto'].value_counts().index[0:5])

ax.set_title('Top 5 Compostos', fontweight='bold', fontsize='15')
plt.xlabel('Composto', fontweight='light', fontsize='14')
plt.ylabel('Quantidade Usada', fontweight='light', fontsize='14')

plt.show()
```

###Desafio 03:


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
sns.histplot(data = dados, x='g-7', kde=True, bins = 50)

plt.show()
```

###Desafio 04:

<p>O comando ".describe()" traz uma série de dados estatísticos resumindo os dados fornecidos.</p>


*   count: contagem de linhas
*   mean: cálculo da média
*   std: desvio padrão
*   min: valor mínimo
*   25%, 50% e 75%: os quartis 
*   max: valor máximo



###Desafio 05:

<p>Com uma boa disposição dos dados é possível ter uma melhor visualição do gráfico e melhor entendimento dos dados fornecidos.</p>
<p>Um exemplo é plotarmos um gráfico com muitas informações em uma área pequena, teremos dados aglomerados, impossibilitando a visualização de informações. Então sempre devemos cuidar do tamanho da visualização e de suas manipulações.</p>
